### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

### Which pokemon has the highest combined value of Attack and Sp. Atk ?

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
#your code here
# Filter Generation 3, non-legendary Pokémon
gen3_non_legendary = data[(data['Generation'] == 3) & (~data['Legendary'])]

# Find the weakest Generation 6, legendary Pokémon
weakest_gen6_legendary_HP = data[(data['Generation'] == 6) & (data['Legendary'])]['HP'].min()

# Count Generation 3, non-legendary Pokémon with higher HP
count = (gen3_non_legendary['HP'] > weakest_gen6_legendary_HP).sum()

print("Number of Generation 3, non-legendary Pokémon with higher HP than the weakest Generation 6, legendary Pokémon:", count)
```

### Print out the third slowest pokemon(s) in the dataset.

```python
#your code here
# Sort the DataFrame by 'Speed' in ascending order
sorted_data = data.sort_values(by='Speed')

# Find the third slowest speed
third_slowest_speed = sorted_data['Speed'].iloc[2]

# Filter Pokémon with the third slowest speed
third_slowest_pokemon = sorted_data[sorted_data['Speed'] == third_slowest_speed]

# Print out the third slowest Pokémon(s)
print("Third slowest Pokémon(s):")
print(third_slowest_pokemon)
```

### Which pokemon type has the highest average HP?

```python
#your code here
# Group the data by 'Type 1' and calculate the average HP for each type
average_hp_by_type = data.groupby('Type 1')['HP'].mean()

# Find the type with the highest average HP
highest_avg_hp_type = average_hp_by_type.idxmax()
highest_avg_hp_value = average_hp_by_type.max()

print("Type with the highest average HP:", highest_avg_hp_type)
print("Average HP:", highest_avg_hp_value)
```
